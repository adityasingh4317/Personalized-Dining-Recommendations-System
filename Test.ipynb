```python
def detect_language(text):
    try:
        lang = detect(text)
        return lang == 'en'  # Return True if language is English, False otherwise
    except:
        return False  # Return False if an error occurs during language detection
```

```python
def calculate_wait_time(customer_order, df):
    total_wait_time = 0
    for dish in customer_order:
        selected_dish = df.loc[df['name_of_Dish'] == dish]
        if not selected_dish.empty:
            prep_time = selected_dish.iloc[0]['Prepration_time']
            cook_time = selected_dish.iloc[0]['Cooking_time']
            total_wait_time += prep_time + cook_time
        else:
            print(f"Dish '{dish}' not found in the menu.")
    return total_wait_time

# Prompt the user to input their order
user_input = input("Enter the dishes you want to order (separated by commas): ")
customer_order = [dish.strip() for dish in user_input.split(',')]
print("Dish Name:", user_input)
total_wait_time = calculate_wait_time(customer_order, df)
print(f"Total estimated wait time for the order: {total_wait_time} minutes")
```

```python
# Function to recommend dishes based on the Similar_Dishes column
def recommend_dishes(customer_order, df):
    recommended_dishes = []
    for dish in customer_order:
        similar_dishes = df[df['name_of_Dish'] == dish.strip()]['Similar_Dishes'].tolist()
        if similar_dishes:
            recommended_dishes.extend(similar_dishes[0].split(', '))
    return recommended_dishes

# Simulate testing scenarios with user input
def user_input_simulation(df):
    num_customers = int(input("Enter the number of customers: "))
    test_results = []

    for i in range(num_customers):
        print(f"\nCustomer {i + 1}")
        customer_order = input("Enter the dishes you want to order (comma-separated): ").strip().split(',')
        expected_wait_time = int(input("Enter the expected wait time for your order (in minutes): "))
        satisfaction_rating = int(input("Enter your satisfaction rating (1-10): "))

        # Calculate wait time
        actual_wait_time = calculate_wait_time(customer_order, df)

        # Recommend dishes based on Similar_Dishes column
        recommended_dishes = recommend_dishes(customer_order, df)

        # Feedback on wait time estimation
        if actual_wait_time == expected_wait_time:
            feedback = "Wait time estimation accurate."
        else:
            feedback = f"Wait time estimation inaccurate. Expected: {expected_wait_time}, Actual: {actual_wait_time}"

        # Store the test result
        test_result = {
            'Customer_Order': customer_order,
            'Expected_Wait_Time': expected_wait_time,
            'Actual_Wait_Time': actual_wait_time,
            'Recommended_Dishes': recommended_dishes,
            'Satisfaction_Rating': satisfaction_rating,
            'Feedback': feedback
        }
        test_results.append(test_result)

        # Display test results and feedback
        print(f"\nOrder: {customer_order}")
        print(f"Expected Wait Time: {expected_wait_time} minutes")
        print(f"Actual Wait Time: {actual_wait_time} minutes")
        print(f"Recommended Dishes: {recommended_dishes}")
        print(f"Satisfaction Rating: {satisfaction_rating}")
        print(f"Feedback: {feedback}")

    return test_results

# Run the user input simulation
test_results = user_input_simulation(df)

# Display all test results (optional)
print("\nAll Test Results:")
for result in test_results:
    print(result)
```
